# Appendix: Resources

The papers, posts, and docs this course leans on — worth reading once the fundamentals click. Mirrors n8n's Resources appendix. (No code here; it's a reading list.)

## Foundational reading

- **Anthropic — [Building Effective Agents](https://www.anthropic.com/engineering/building-effective-agents)** (2024). The best short read in the field, and the source of the workflow patterns (0.4) and the "reach for the simplest thing" discipline that runs through this course.
- **Anthropic — [Writing Effective Tools for Agents](https://www.anthropic.com/engineering/writing-tools-for-agents)**. Why the tool description is the number-one lever on agent quality (1.2).
- **ReAct: Synergizing Reasoning and Acting in Language Models** — Yao et al., 2022 ([arxiv 2210.03629](https://arxiv.org/abs/2210.03629)). The reason-act-observe loop you built by hand in 0.3.
- **Self-Refine: Iterative Refinement with Self-Feedback** — Madaan et al., 2023 ([arxiv 2303.17651](https://arxiv.org/abs/2303.17651)). The generate → critique → improve loop behind reflection (0.5, 2.5).


## Courses

- **[UC Berkeley — Agentic AI](https://agenticai-learning.org/)** (Dawn Song). The pillars this course maps to (0.0), and a deep treatment of evaluation and safety.
- **The n8n no-code course** ([ezponda.github.io/ai-agents-course](https://ezponda.github.io/ai-agents-course/)) — this course's sibling; every chapter here cross-references it.
