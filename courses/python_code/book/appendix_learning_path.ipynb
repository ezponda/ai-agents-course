# Appendix: The AI Agents Engineer Path

You finished the course: you can write the loop from scratch, ship a typed agent with PydanticAI, orchestrate it with LangGraph, and put it behind an API. This appendix answers the next question — *how do I turn that into a job building agents?*

It has three parts: what companies actually evaluate, what to build to prove it, and which external courses are worth your time (and which to skip).

```{note}
The course landscape section below is a snapshot of **mid-2026**. Courses rotate every semester — always check the linked pages for the current edition.
```

## What companies actually evaluate

"Knows LangGraph" is not a hiring bar. Teams that ship agents look for a specific mix:

| Competency | What it means in practice |
|---|---|
| Software engineering | Python, APIs, tests, CI, Docker, deploy, observability |
| Product sense | When to use an LLM call, a workflow, an agent loop, or multi-agent — and when **not** to use agents |
| Tool design | Clear contracts, least privilege, testable, observable |
| Context engineering | Prompts, memory, RAG, history compaction, token budgets |
| Evals | Datasets, trajectory evals, LLM-as-judge, regression suites, failure analysis |
| Reliability | Retries, timeouts, idempotency, concurrency, checkpointing |
| Security | Permissions, secrets, sandboxing, prompt injection, human approval |
| Production | Auth, rate limiting, tracing, cost tracking, incident review |

**The differentiator is evals.** Most candidates can wire up tools; few can show a versioned eval dataset, explain a trajectory eval, or walk through a failure analysis. If you invest disproportionately anywhere, invest there — this course's evals chapters (0.5, 16, 29) are the starting point, not the finish line.

## The interview loop

The typical loop at an agent-focused company in 2026 has four or five rounds:

1. **Coding** — classic algorithms/practical coding, sometimes explicitly *with* an AI assistant, where they evaluate how you direct it. Practice both modes.
2. **Agent system design** — "design a support agent for product X at scale." This is the differential round. Structure your answer: requirements, workflow vs. agent, tools and contracts, context strategy, evals, failure modes and security, cost and latency, rollout.
3. **Debugging / evals practical** — you get traces from a failing agent and diagnose it, or design the eval for a feature.
4. **Project deep dive** — 45-60 minutes on something you built. Numbers win this round: pass rates, cost per run, latency, failure categories, what you changed after the postmortem.
5. **Behavioral** — shipping stories, trade-offs, incidents.

::::{dropdown} 🛠️ Try it: a system-design drill
:color: secondary

Pick one prompt below and answer it out loud, timed, 40 minutes, following the structure above:

- Design a customer-support agent for a bank (escalation, PII, audit).
- Design an internal data-analyst agent over a warehouse (sandboxing, cost caps).
- Design a coding agent for a large monorepo (context, verification, permissions).
- Design an enterprise assistant that connects to five internal systems via MCP (scopes, audit logs).

Then write down which questions you could not answer concretely — each one is a gap to close with a build, not with more reading.
::::

## Rule number one: build outside the notebook

Courses — including this one — get you to "I understand agents." Interviews are won at "I built, deployed, evaluated, and broke one, and here are the numbers." Pick **one** project and take it unreasonably far, rather than three demos.

Project shapes that map well to real teams:

| Project | Why it works |
|---|---|
| **Data analyst agent with a sandbox** | Code execution + security + verifiable answers. Chapter 32 is the embryo |
| **Coding agent + benchmark** | Tests pass or they don't — the cleanest verification story. Chapter 07 is the base |
| **Support agent with real policies** | RAG, routing, human approval, injection defense — closest to what agent companies sell |
| **Enterprise assistant over MCP** | Your own MCP server, scoped tools, audit logs, groundedness evals |
| **Operations workflow agent** | Deterministic workflow + LLM only where judgment is needed — proves you know when *not* to agent |

**What "unreasonably far" means:** real auth with separated users and conversations, Postgres for state and checkpoints, a real sandbox for code execution, per-run traces (user, model, tools, latency, cost), a versioned eval dataset wired into CI, and a written postmortem of 5-10 real failures.

**Portfolio checklist** — every serious project should have: a README with the problem, user, and architecture diagram; a deployed demo or short video; the tool list with contracts; a minimal threat model; eval results (pass rate, cost, latency); unit and integration tests in CI; structured logs or tracing; the postmortem; and a "what I would do next in production" section.

## Common mistakes

- Learning frameworks without being able to write the loop by hand (you are past this — keep it that way).
- Three half-finished demos instead of one project you can defend for an hour.
- Measuring only final answers, never trajectories.
- `exec()` or raw shell without a sandbox, described as "production".
- One shared API key, described as "auth".
- No cost tracking, no concurrency testing, no regression evals.
- Building multi-agent because it sounds advanced.
- Walking into a system-design round without ever having practiced one **out loud**.

## The course landscape (as of mid-2026)

Where external courses fit *after* this one — verified July 2026:

| Course | Status | Verdict |
|---|---|---|
| [Berkeley MOOC series](https://agenticai-learning.org/) (LLM Agents F24, Advanced Sp25, Agentic AI F25) | Ended; certificate tracks closed; lectures and slides free on [YouTube](https://www.youtube.com/@BerkeleyRDI) | Audit selectively. Best single lecture for practitioners: Sierra's "Practical Lessons from Deploying Real-World AI Agents" (F25). Watch the site for new editions |
| [CMU 11-768 AI Agents](https://www.cmu-agents.com/) (Neubig & Fried) | Starts Fall 2026 | The first build-focused university agents course: agent harnesses, agent evals, agentic RL. Follow the public materials as they appear |
| [Stanford CS224G](https://web.stanford.edu/class/cs224g/schedule.html) | Winter editions; materials public | Use its schedule as a checklist against your portfolio project |
| [Evaluating AI Agents](https://www.deeplearning.ai/courses/evaluating-ai-agents) (DeepLearning.AI + Arize) | Active, free to audit | Do it — the best short eval course |
| [LangChain Academy](https://academy.langchain.com/) | Active, free | *Introduction to Deep Agents* and *Ambient Agents* are the two worth your time after Block 2 |
| [Anthropic Academy](https://anthropic.skilljar.com/) | Active, free, certificates | The two MCP courses and *Building with the Claude API* complement chapters 18-18b |
| [OpenAI Agents SDK docs](https://openai.github.io/openai-agents-python/) | Docs, not a course | Do the quickstart + handoffs + guardrails; labs expect you to know their native SDKs |
| [Hugging Face Agents Course](https://huggingface.co/learn/agents-course/unit0/introduction) | Maintained, frameworks frozen in 2025 | Skip if you did this course — heavy overlap |

**Rule of thumb:** one build course (this one), one eval course, the native SDKs of the labs you target, and selective lectures for research literacy. More courses than that is procrastination with a syllabus.

## Essential readings

Short list, all load-bearing — these come up in interviews:

- Anthropic, [Building Effective Agents](https://www.anthropic.com/engineering/building-effective-agents) — workflows vs. agents; start simple.
- Anthropic, [Demystifying evals for AI agents](https://www.anthropic.com/engineering/demystifying-evals-for-ai-agents) — pairs directly with chapters 16 and 29.
- Anthropic, [How we contain Claude across products](https://www.anthropic.com/engineering/how-we-contain-claude) — sandboxing and blast radius, the security side of tool design.
- Anthropic, [Harness design for long-running application development](https://www.anthropic.com/engineering/harness-design-long-running-apps) and [Scaling Managed Agents](https://www.anthropic.com/engineering/managed-agents).
- OpenAI, [A practical guide to building agents](https://cdn.openai.com/business-guides-and-resources/a-practical-guide-to-building-agents.pdf) (PDF) and the [Agents SDK](https://openai.github.io/openai-agents-python/).
- Sierra, [Meet the AI Agent Engineer](https://sierra.ai/blog/meet-the-ai-agent-engineer) — the clearest description of the role itself. Read their job posts as a spec.
- [MCP documentation](https://modelcontextprotocol.io/docs/getting-started/intro).

The ideas they all repeat: the building block is the augmented LLM, not the framework; single-agent before multi-agent; evals before optimization; the distance from demo to production is the job.

## Where this leaves you

The profile that passes the filter is not "I took courses on agents." It is:

> I built and deployed an agentic system with real tools, permissions, a sandbox, tracing, versioned evals in CI, idempotent effects, and human-in-the-loop. I have the numbers and the postmortems. And I can tell you when an agent is the wrong answer.

Everything in this course points at that sentence. The gap between finishing these notebooks and saying it truthfully is one serious project — go build it.
