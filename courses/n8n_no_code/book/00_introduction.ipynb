# Introduction to n8n

n8n is a visual workflow automation tool. You connect nodes on a canvas, and data flows through them. No code required.

This course teaches you to build AI workflows—automations that call language models like GPT to process text, make decisions, and generate responses.

**By the end, you will have built:**
- A workflow that writes a memo in three steps (outline → edit → final draft)
- A routing system that classifies support tickets and sends each to the right handler
- An AI agent that can search Wikipedia, do math, and remember your conversation

All without writing a single line of code.

---

## Prerequisites

- **Docker Desktop** (recommended) or Node.js 18+
- A browser to access the n8n interface at `localhost:5678`
- An API key from OpenRouter, OpenAI, or Google AI (you'll set this up in Quick Start)

---

## How n8n Works

You build workflows by placing **nodes** on a canvas and connecting them with lines. Each node does one thing: fetch data, call an API, transform text, make a decision.

```
[Trigger] → [Get Data] → [Call AI] → [Save Result]
```

Data flows left to right. When you run the workflow, each node executes in sequence.

n8n can run on your computer (self-hosted) or in the cloud. This course uses the self-hosted version so you can experiment freely.

---

## Key Terms

A few terms you'll see throughout the course:

| Term | Meaning |
|------|---------|
| **Workflow** | A sequence of connected nodes that automates a task |
| **Node** | One step in a workflow (e.g., "call GPT", "save to file") |
| **Trigger** | The first node—defines what starts the workflow |
| **Execution** | One complete run of a workflow |

You'll learn these by doing. The Quick Start chapter has you building within the first few minutes.

## A Note on Costs

Some workflows in this course call paid AI APIs (OpenRouter, OpenAI, Google). To avoid unexpected charges:

- Run workflows in **Test mode** (manual execution) while learning
- Keep workflows **deactivated** until you're ready for production
- Tools like Wikipedia and Calculator are free—no API key needed
